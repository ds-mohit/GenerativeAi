### 1. How to load text file

```python
from langchain_community.document_loaders import TextLoader
```

```python
# this text loader is a class so we have to create a object
loader=TextLoader("speech.txt")
loader # this will print object

```

```python
# to see the data 
text_doc=loader.load()
text_doc
```

### 2.  How to load a pdf file

```python
from langchain_community.document_loaders import PyPDFLoader
```

```python
loader_pdf=PyPDFLoader("attention.pdf")
pdf_doc=loader_pdf.load()
pdf_doc
```

### 3. How to load live web page data (webbase loader)

```python
from langchain_community.document_loaders import WebBaseLoader # this is used to load the data from live sites 
## that allow us to see the data

loader_web=WebBaseLoader("https://lilianweng.github.io/")
web_doc=loader_web.load()
web_doc
```

### 4. Loading data from wikipedia

```python
from langchain_community.document_loaders import WebBaseLoader

loader_wiki=WebBaseLoader("https://en.wikipedia.org/wiki/Generative_artificial_intelligence")
web_doc=loader_wiki.load()
web_doc
```
